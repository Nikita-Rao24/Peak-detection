### Load and visualise different audio files 
Here we use Fast Fourier transform to visualise high peaks in the audio.
For first two samples we use librosa's waveplot to visualise  the peaks.
For the rest of the audio file we use matplotlib to visualise the peaks.

```python
#class car crash
import librosa
audio_data = r'C:/Users/nikita.ramesh.rao/Desktop/test/sample/fold5/car_crash.wav'
x , sr = librosa.load(audio_data)
import IPython.display as ipd
ipd.Audio(audio_data)
```

```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
fft_out = fft(x)
%matplotlib inline
_ = librosa.display.waveplot(np.abs(fft_out),sr=sr)
```
